```python
#ist a python file to recogrize a Least Recently Used

from collections import deque
class LRU(object):
    def __init__(self, pageNumber):
        self.pageNumber = pageNumber
        self.queue = deque()
        self.hashSet = set()
    def appendPage(self, page):
        if self.pageNumber == len(self.queue):
            self.hashSet.remove(self.queue[-1])
            self.queue.pop()
        self.hashSet.add(page)
        self.queue.appendleft(page)
    def accessPage(self, page):
        if page not in self.hashSet:
            self.appendPage(page)
        elif page != self.queue[0]:
            self.queue.remove(page)
            self.queue.appendleft(page)
    def getQueue(self):
        return self.queue
object = LRU(4)

object.accessPage(1)
object.accessPage(2)
object.accessPage(3)
object.accessPage(4)
object.accessPage(5)
object.accessPage(6)

object.getQueue()
       
```

```python
# judge whether two listnode similar or not

def isSimilar(head1, head2):
    cur1, cur2 = head1, cur2
    hashSet = set()
    while cur1:
        hashSet.add(cur1)
        cur1 = cur1.next
    while cur2:
        if cur2 in hashSet:
            return True
        cur2 = cur2.next
    return False
            
```
